```python
import pandas as pd
import matplotlib.pyplot as plt

# Veri setini oku
df = pd.read_csv("netflix_titles.csv")

# 1) Yıllara göre içerik sayısı
yearly_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,5))
yearly_counts.plot(kind='line')
plt.title("Yıllara Göre Netflix İçerik Sayısı")
plt.xlabel("Yıl")
plt.ylabel("İçerik Sayısı")
plt.show()

# 2) Film vs TV Show
type_counts = df['type'].value_counts()
plt.figure(figsize=(6,6))
type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Film ve TV Show Dağılımı")
plt.ylabel("")
plt.show()

# 3) En çok içerik üreten 10 ülke
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(10,5))
country_counts.plot(kind='bar')
plt.title("En Çok İçerik Üreten Ülkeler (Top 10)")
plt.xlabel("Ülke")
plt.ylabel("İçerik Sayısı")
plt.show()
```
